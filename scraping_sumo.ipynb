```python

df = df_0.copy()
# property_name
df['property_name'] = df['property_name'].str[2:]
# rent
df['rent'] = (df['rent'].str[:-2].astype(float) * 10000).astype(int)
# management_fee
df['management_fee'] = df['management_fee'].str[:-1]
df.loc[df['management_fee']=='','management_fee'] = 0
df['management_fee'] = df['management_fee'].astype(int)
# deposit
df['deposit'] = df['deposit'].str[:-2]
df.loc[df['deposit']=='','deposit'] = 0
df['deposit'] = (df['deposit'].astype(float)*10000).astype(int)
# key_money
df['key_money'] = df['key_money'].str[:-2]
df.loc[df['key_money']=='','key_money'] = 0
df['key_money'] = (df['key_money'].astype(float)*10000).astype(int)
# security_deposit
df['security_deposit'] = df['security_deposit'].str[:-2]
df.loc[df['security_deposit']=='','security_deposit'] = 0
df['security_deposit'] = (df['security_deposit'].astype(float)*10000).astype(int)
# discount
df['discount'] = df['discount'].str[:-2]
df.loc[df['discount']=='','discount'] = 0
df['discount'] = (df['discount'].astype(float)*10000).astype(int)
# floor_plan
df['occupied_area'] = df['occupied_area'].str[:-2]
df.loc[df['occupied_area']=='','occupied_area'] = 0
df['occupied_area'] = df['occupied_area'].astype(float)
df = df.rename(columns={'occupied_area':'occupied_area_m^2'})
# year_build
df['year_build'] = df['year_build'].str[1:-1]
df.loc[df['year_build']=='','year_build'] = 0
df['year_build'] = df['year_build'].astype(int)
# access
df['access'] = df['access'].str.split('\n')
# feature_facility
df['feature_facility'] = df['feature_facility'].str.split('、')
# floor_plan_detail
df['floor_plan_detail'] = df['floor_plan_detail'].str.split(' ')
df['floor_plan_detail_type'] = df['floor_plan_detail'].apply(lambda x: x[0][:1])
df['floor_plan_detail_m^2'] = df['floor_plan_detail'].apply(lambda x: x[0][1:])
df.loc[df['floor_plan_detail_m^2']=='','floor_plan_detail_m^2'] = 0
df['floor_plan_detail_m^2'] = df['floor_plan_detail_m^2'].astype(float)
df.loc[df['floor_plan_detail_m^2']==0,'floor_plan_detail_m^2'] = np.NAN
df['floor_plan_detail_loft'] = df['floor_plan_detail'].apply(lambda x: len(x))
loft = df.loc[df['floor_plan_detail_loft']==3,'floor_plan_detail'].apply(lambda x: x[2])
df.loc[df['floor_plan_detail_loft']!=3,'floor_plan_detail_loft'] = 0
df.loc[df['floor_plan_detail_loft']==3,'floor_plan_detail_loft'] = loft
df['floor_plan_detail_loft'] = df['floor_plan_detail_loft'].astype(float)
df.loc[df['floor_plan_detail_loft']==0,'floor_plan_detail_loft'] = np.NAN
df.drop(columns='floor_plan_detail',inplace=True)
# structure
df['storey'] = df['storey'].apply(lambda x: x.split('/'))
df['max_storey'] = df['storey'].apply(lambda x: x[1])
df['storey'] = df['storey'].apply(lambda x: x[0])
#date_build
df['build_year'] = df['date_build'].apply(lambda x: x[:x.find('年')]).astype(int)
df['build_month'] = df['date_build'].apply(lambda x: x[x.find('年')+1:x.find('月')]).astype(int)
df.drop(columns='date_build',inplace=True)
# condition
df['condition'] = df['condition'].apply(lambda x: x.split('/'))
# households
df['households'] = df['households'].str[:-1]
df['households'] = df['households'].replace('',0)
df['households'] = df['households'].astype(int)
# update_date
df['update_date']  = pd.to_datetime(df['update_date']) 


```
